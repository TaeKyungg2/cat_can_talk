```python
import pandas as pd
import numpy as np

data=pd.read_csv('sample_sentences.csv')
data=data.T
data.head(10)
```

```python
data.head(50)
```

```python
import re

def change_to_cat_style(sentence):
    sentence.strip()
    sentence=re.sub(r'다$|요$|죠$|지$',sentence+'냥',sentence)
    sentence=re.sub(r'습니다$|습니까$|나요$',sentence+'냥?',sentence)
    sentence=re.sub(r'양|얌|어요$|나$|이$','냥',sentence)
    sentence=re.sub(r'지만|,',r'\g<0> ,냐옹, ',sentence)
    sentence=re.sub(r'!+',r'\g<0>냐아아옹!',sentence)
    return sentence
```

```python
trans_said=data.map(change_to_cat_style)
```
